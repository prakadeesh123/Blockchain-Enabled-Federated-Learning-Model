```python
import hashlib
import json
import time

class Blockchain:
    def __init__(self):
        self.chain = []
        self.transactions = []
        self.create_block(proof=1, previous_hash="0")  # Genesis block

    def create_block(self, proof, previous_hash):
        block = {
            'index': len(self.chain) + 1,
            'timestamp': time.time(),
            'transactions': self.transactions,
            'proof': proof,
            'previous_hash': previous_hash
        }
        self.transactions = []  # Reset transactions
        self.chain.append(block)
        return block

    def get_previous_block(self):
        return self.chain[-1]

    def proof_of_work(self, previous_proof):
        new_proof = 1
        while True:
            hash_value = hashlib.sha256(str(new_proof**2 - previous_proof**2).encode()).hexdigest()
            if hash_value[:4] == "0000":  # PoW condition
                return new_proof
            new_proof += 1

    def hash(self, block):
        return hashlib.sha256(json.dumps(block, sort_keys=True).encode()).hexdigest()

    def add_transaction(self, sender, receiver, model_hash):
        self.transactions.append({'sender': sender, 'receiver': receiver, 'model_hash': model_hash})
        return self.get_previous_block()['index'] + 1
```

```python
from Crypto.Cipher import AES
import base64

key = b'16char_secretkey'  # Must be 16, 24, or 32 bytes

def encrypt_data(data):
    cipher = AES.new(key, AES.MODE_EAX)
    nonce = cipher.nonce
    ciphertext, tag = cipher.encrypt_and_digest(data.encode())
    return base64.b64encode(nonce + ciphertext).decode()

def decrypt_data(encrypted_data):
    encrypted_data = base64.b64decode(encrypted_data)
    nonce = encrypted_data[:16]
    ciphertext = encrypted_data[16:]
    cipher = AES.new(key, AES.MODE_EAX, nonce=nonce)
    return cipher.decrypt(ciphertext).decode()
```

```python
def average_model_weights(model_paths):
    all_weights = []

    for path in model_paths:
        model = tf.keras.models.load_model(path, compile=False)
        all_weights.append(model.get_weights())

    # Only keep layers with matching shapes across all models
    common_weights = []
    for layers in zip(*all_weights):
        if all(w.shape == layers[0].shape for w in layers):
            avg = np.mean(np.array(layers), axis=0)
            common_weights.append(avg)
        else:
            print("Skipping mismatched layer")

    return common_weights
```

```python
def build_global_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(256, activation='relu', input_shape=(input_shape,)),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
import os
import pandas as pd
import numpy as np
import joblib
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf

test_csv = '/content/drive/MyDrive/testdataset.csv'
local_model_base_path = '/content/drive/MyDrive/keras'


test_data = pd.read_csv(test_csv)
y_test = test_data['label']
X_test_raw = test_data.drop(columns=['label'])
X_test_numeric = X_test_raw.select_dtypes(include=[np.number])


for client_folder in sorted(os.listdir(local_model_base_path)):
    client_path = os.path.join(local_model_base_path, client_folder)

    if not os.path.isdir(client_path):
        continue

    try:
        print(f"\nEvaluating model from: {client_folder}")


        scaler_path = os.path.join(client_path, f'scaler_client1.pkl')
        scaler = joblib.load(scaler_path)


        expected_columns = scaler.feature_names_in_
        common_columns = list(set(expected_columns).intersection(X_test_numeric.columns))
        X_test = X_test_numeric[common_columns]
        X_test = X_test.reindex(columns=expected_columns, fill_value=0)
        X_test_scaled = scaler.transform(X_test)

        model_path = os.path.join(client_path, f'{client_folder.replace("_results", "")}_weights.h5')
        model = tf.keras.models.load_model(model_path)


        y_pred_probs = model.predict(X_test_scaled)
        y_pred = (y_pred_probs > 0.5).astype(int).flatten()

        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))

        print("\nClassification Report:")
        print(classification_report(y_test, y_pred, digits=4))

    except Exception as e:
        print(f"‚ùå Error with {client_folder}: {e}")
```

```python
import os
import requests
import tempfile
import numpy as np
import tensorflow as tf
from web3 import Web3
from keras.models import load_model
from keras.saving import save_model
from scipy.optimize import minimize
import warnings
warnings.filterwarnings("ignore")

infura_url = "https://sepolia.infura.io/v3/e89eff990e004123b9d421a585f34a4b"
web3 = Web3(Web3.HTTPProvider(infura_url))

contract_address = Web3.to_checksum_address("0x7888d98a7533f6e4e7881cd891ae8a653f434e4b")
contract_abi = [
    {"inputs": [], "name": "getTotalUpdates", "outputs": [{"internalType": "uint256", "name": "", "type": "uint256"}], "stateMutability": "view", "type": "function"},
    {"inputs": [{"internalType": "uint256", "name": "index", "type": "uint256"}],
     "name": "getUpdate",
     "outputs": [
         {"internalType": "address", "name": "", "type": "address"},
         {"internalType": "string", "name": "", "type": "string"},
         {"internalType": "uint256", "name": "", "type": "uint256"}
     ],
     "stateMutability": "view", "type": "function"}
]
contract = web3.eth.contract(address=contract_address, abi=contract_abi)

def download_ipfs_model(ipfs_hash):
    url = f"https://gateway.pinata.cloud/ipfs/{ipfs_hash}"
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception("IPFS fetch failed")

    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".h5")
    temp_file.write(response.content)
    temp_file.close()
    return temp_file.name

def fetch_valid_models():
    valid_model_paths = []
    total = contract.functions.getTotalUpdates().call()
    for i in range(total):
        try:
            _, ipfs_hash, _ = contract.functions.getUpdate(i).call()
            if ipfs_hash.endswith('.json'):
                print(f"Skipping non-model file (json): {ipfs_hash}")
                continue
            path = download_ipfs_model(ipfs_hash)
            _ = load_model(path, compile=False)
            print(f"‚úÖ Model {i} loaded: {ipfs_hash}")
            valid_model_paths.append(path)
        except Exception as e:
            print(f"‚ùå Skipping model {i}: {e}")
    return valid_model_paths

def geometric_median(points, eps=1e-5):
    def aggregate(x):
        return sum(np.linalg.norm(x - p) for p in points)
    x0 = np.mean(points, axis=0)
    res = minimize(aggregate, x0, method='COBYLA')
    return res.x

def geometric_median_averaging(model_paths):
    models = [load_model(p, compile=False) for p in model_paths]
    base_model = models[0]
    base_weights = base_model.get_weights()

    averaged_weights = []
    for i, ref_weight in enumerate(base_weights):
        compatible = []
        for model in models:
            w = model.get_weights()
            if len(w) > i and w[i].shape == ref_weight.shape:
                compatible.append(w[i])

        if not compatible:
            print(f"‚ùå Skipping layer {i}: no shape match.")
            averaged_weights.append(ref_weight)
        elif len(ref_weight.shape) > 1:
            avg_tensor = np.median(compatible, axis=0)
            averaged_weights.append(avg_tensor)
        else:
            avg_vector = geometric_median(np.array(compatible))
            averaged_weights.append(avg_vector)

    return averaged_weights, base_model

model_paths = fetch_valid_models()
averaged_weights, base_model = geometric_median_averaging(model_paths)

base_model.set_weights(averaged_weights)
save_path = "/content/global_model_geomedian1.h5"
base_model.save(save_path)
print(f"\n‚úÖ Final global model saved using geometric median: {save_path}")
```

```python
import os
import requests
import tempfile
import numpy as np
import tensorflow as tf
from web3 import Web3
from keras.models import load_model
from keras.saving import save_model
model_accuracies = {
    "client1_results": 0.9781,
    "client2_results": 0.9819,
    "client3_results": 0.9799,
    "client4_results": 0.9703,
}

infura_url = "https://sepolia.infura.io/v3/e89eff990e004123b9d421a585f34a4b"
web3 = Web3(Web3.HTTPProvider(infura_url))

contract_address = Web3.to_checksum_address("0x7888d98a7533f6e4e7881cd891ae8a653f434e4b")
contract_abi = [
    {"inputs": [], "name": "getTotalUpdates", "outputs": [{"internalType": "uint256", "name": "", "type": "uint256"}], "stateMutability": "view", "type": "function"},
    {"inputs": [{"internalType": "uint256", "name": "index", "type": "uint256"}],
     "name": "getUpdate",
     "outputs": [
         {"internalType": "address", "name": "", "type": "address"},
         {"internalType": "string", "name": "", "type": "string"},
         {"internalType": "uint256", "name": "", "type": "uint256"}
     ],
     "stateMutability": "view", "type": "function"}
]
contract = web3.eth.contract(address=contract_address, abi=contract_abi)

model_accuracies = {
    "QmakPTdF7yRHTNnLfZ4tugznvEJ7LhsjJGeYv6jQDfXZtP": 0.9781,  # client1
    "Qmc4FDpDGtAKNNnsA754hXrmhiJqcsJtbZtQd2dsM24L99": 0.9819,  # client2
    "QmZ1qbnazE6rdVGKqMZa1FaTZDLQUiPyRePCTLTNwZ3kmB": 0.9799,  # client3
    "QmRN3U7pYYYDrx4MKPiiqtTeiwVAM4StRbFmeFQVV3rfDk": 0.9703,  # client4
}

def download_ipfs_model(ipfs_hash):
    url = f"https://gateway.pinata.cloud/ipfs/{ipfs_hash}"
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception("IPFS fetch failed")

    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".h5")
    temp_file.write(response.content)
    temp_file.close()
    return temp_file.name


def fetch_models_with_accuracy():
    total = contract.functions.getTotalUpdates().call()
    model_paths = []
    weights = []
    for i in range(total):
        try:
            _, ipfs_hash, _ = contract.functions.getUpdate(i).call()
            if ipfs_hash.endswith(".json") or ipfs_hash not in model_accuracies:
                continue
            path = download_ipfs_model(ipfs_hash)
            model = load_model(path, compile=False)
            model_paths.append(model)
            weights.append(model_accuracies[ipfs_hash])
            print(f"‚úÖ Model {i} loaded: {ipfs_hash}")
        except Exception as e:
            print(f"‚ùå Model {i} skipped: {e}")
    return model_paths, weights


def accuracy_weighted_average(models, weights):
    weights = np.array(weights)
    weights = weights / np.sum(weights)  # normalize

    base_weights = models[0].get_weights()
    averaged = []

    for i in range(len(base_weights)):
        compatible_weights = [m.get_weights()[i] for m in models if len(m.get_weights()) > i and m.get_weights()[i].shape == base_weights[i].shape]
        if len(compatible_weights) != len(models):
            print(f"Layer {i} updated.")
            averaged.append(base_weights[i])
            continue
        stacked = np.stack(compatible_weights, axis=0)
        avg = np.tensordot(weights, stacked, axes=1)
        averaged.append(avg)
    return averaged
models, acc_weights = fetch_models_with_accuracy()
final_weights = accuracy_weighted_average(models, acc_weights)
global_model = models[0]
global_model.set_weights(final_weights)
global_model.save("/content/global_model_weighted_avg.h5")
print("‚úÖ Global model saved using accuracy-based weighted averaging.")
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.metrics import confusion_matrix, classification_report
import tensorflow as tf


test_data = pd.read_csv('/content/drive/MyDrive/testdataset.csv')

y_test = test_data['label']

X_test_raw = test_data.drop(columns=['label'])

X_test_numeric = X_test_raw.select_dtypes(include=[np.number])

scaler = joblib.load('/content/drive/MyDrive/keras/client4_results/scaler_client1.pkl')
model = tf.keras.models.load_model('/content/drive/MyDrive/final_global_model.h5')

expected_columns = scaler.feature_names_in_

common_columns = list(set(expected_columns).intersection(X_test_numeric.columns))

X_test = X_test_numeric[common_columns]
X_test = X_test.reindex(columns=expected_columns, fill_value=0)

X_test_scaled = scaler.transform(X_test)


y_pred_probs = model.predict(X_test_scaled)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```
